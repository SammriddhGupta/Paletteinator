```python
van_gogh_image = plt.imread('ghw-2022-09/art/Vincent_van_Gogh/Vincent_van_Gogh_385.jpg')
plt.imshow(van_gogh_image)
```

```python
def posterize_pixel(pixel):
  new_red_value = ((pixel[0] // 64) * 64) + 32
  new_green_value = ((pixel[1] // 64) * 64) + 32
  new_blue_value = ((pixel[2] // 64) * 64) + 32

  return (new_red_value, new_green_value, new_blue_value)
```

```python
def posterize_image(image):
  new_image = []
  for row in image:
    new_row = []
    for pixel in row:
      new_row.append(posterize_pixel(pixel))
    new_image.append(new_row)
  return new_image


plt.imshow(posterize_image(heart))
```

```python
plt.imshow(van_gogh_image)
plt.show()

posterized_van_gogh_image = posterize_image(van_gogh_image)
plt.imshow(posterized_van_gogh_image)
```

```python
def most_common_colors(image, n=3):
  pixels = []
  for row in image:
    for pixel in row:
      pixels.append(tuple(pixel))
  pixel_counter = Counter(pixels)
  most_common_color_tuples = pixel_counter.most_common(n=n)

  most_common_colors = [most_common_color for most_common_color, count in most_common_color_tuples]
  
  return most_common_colors

plt.imshow([most_common_colors(van_gogh_image, n=10)])
```

```python
posterized_most_common_colors = most_common_colors(posterized_van_gogh_image, n=10)
```

```python
artists = ['Vincent_van_Gogh' , 'Rembrandt' , 'Andy_Warhol']
train_image_dict = {artist: list() for artist in artists}

for artist in artists:
  artist_dir = f'ghw-2022-09/art/{artist}'
  first_three_files = os.listdir(artist_dir)[:3]
  for train_image_file in first_three_files:
    train_image = plt.imread(f'{artist_dir}/{train_image_file}')
    train_image_dict[artist].append(train_image)
```

```python
train_image_dict
```

```python
from collections import Counter
from tqdm import tqdm

artist_common_colors = {}

for artist in train_image_dict:
  most_common_colors = Counter()
  for train_image in tqdm(train_image_dict[artist]):
    posterized_image = posterize_image(train_image)
    for row in posterized_image:
      most_common_colors.update(row)
  artist_common_colors[artist] = [color for color, count in most_common_colors.most_common(n=5)]
```

```python
artist_common_colors
```

```python
for artist in artist_common_colors:
  plt.imshow([artist_common_colors[artist]])
  plt.title(artist)
  plt.show()
```

```python
def predict_artist(image, palette_dictionary):
  posterized_image = posterize_image(image)
  #artist_color_matches = {artist:0 for artist in palette_dictionary }
  best_match = None
  best_matched_pixels = 0
  for artist in palette_dictionary:
    matched_pixels = 0
    artists_palette = palette_dictionary[artist]
    for row in posterized_image:
      for pixel in row: 
        if pixel in artists_palette:
          matched_pixels +=1
    if matched_pixels > best_matched_pixels:
      best_match = artist
      best_matched_pixels = matched_pixels

  return best_match
```

```python
num_correct = 0
num_predicted = 0
for artist in train_image_dict:
  for train_image in train_image_dict[artist]:
    prediction = predict_artist(image=train_image, palette_dictionary=artist_common_colors)
    if prediction == artist:
      num_correct+=1
      num_predicted+=1
      print(f'Real Artist:{artist}; Predicted Artist:{prediction}')
```

```python
accuracy = num_correct / num_predicted
accuracy
```

```python
artists = ['Vincent_van_Gogh' , 'Rembrandt' , 'Andy_Warhol']
train_image_dict = {artist: list() for artist in artists}
test_image_dict = {artist: list() for artist in artists}

for artist in artists:
  artist_dir = f'ghw-2022-09/art/{artist}'
  first_three_files = os.listdir(artist_dir)[:3]
  for train_image_file in first_three_files:
    train_image = plt.imread(f'{artist_dir}/{train_image_file}')
    train_image_dict[artist].append(train_image)

for artist in artists:
  artist_dir = f'ghw-2022-09/art/{artist}'
  last_ten_files = os.listdir(artist_dir)[-10:]
  for test_image_file in last_ten_files:
    test_image = plt.imread(f'{artist_dir}/{test_image_file}')
    test_image_dict[artist].append(test_image)
```

```python
num_correct = 0
num_predicted = 0
for artist in test_image_dict:
  for test_image in test_image_dict[artist]:
    prediction = predict_artist(image=test_image, palette_dictionary=artist_common_colors)
    if prediction == artist:
      num_correct+=1
      num_predicted+=1
      print(f'Real Artist:{artist}; Predicted Artist:{prediction}')
```

```python
accuracy = num_correct / num_predicted
accuracy
```
